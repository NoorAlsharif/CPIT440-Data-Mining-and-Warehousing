# 3 - Data visualization


### Histogram

A histogram shows the number of instances (on the vertical axis) that have a given value range (on the horizontal axis). It is a representation of the distribution of data. You can either plot this one attribute at a time, or you can call the `hist()` method on the whole dataset, and it will plot a histogram for each numerical attribute

```python
%matplotlib inline
import matplotlib.pyplot as plt
housing.hist(bins=50, figsize=(20,15))

plt.suptitle('Figure 1', fontsize=20)
plt.show()
```

The `hist()` method relies on `matplotlib` which targets many different formats such as specific graphical user interfaces. In our case, we want the plots to be rendered withen the notebook itself. The magic command `%matplotlib inline` will do this option.  

In the following, we plot the histogram of only one attribute.  

```python
%matplotlib inline
import matplotlib.pyplot as plt
#housing.hist(column='median_income', bins=10, figsize=(20,15))
housing.hist(column='median_income', bins=10)
plt.suptitle('Figure 2', y=1.05, fontsize=20)
plt.show()
```

```python
# To find the intervals of each bin and its count
pd.cut(housing['median_income'], 10).value_counts().sort_index()
# cut convert the continous variable to discrete
```

To understand how to choose `y` argument of `plt.suptitle` look at:  
1. https://stackoverflow.com/questions/55767312/how-to-position-suptitle

**Analysis of Figure 1**  
1. The histogram of many attributes have long tails espetially on the right side, which means that they are positive skewed. For example, in the attribute median_income, it would mean that many districts have income less than the average value, i.e. 3.8. The skewing may make it a bit harder for some Machine Learning algorithms to detect patterns. We will try transforming these attributes later on to have more bell-shaped distributions. 
2. `median_income` attribute does not look like it is expressed in US dollars (USD). The data has been scaled and capped at 15 (actually 15.0001) for higher median incomes, and at 0.5 (actually 0.4999) for lower median incomes. The numbers represent roughly tens of thousands of dollars (e.g., 3 actually means about 30,000).
3. Moreover, we can note that the attributes have very different scales which needs to be processed before going further in the analysis. 
4. The housing median age and the median house value were also capped. The capping is serious problem if we want to build a model that will predect one of the capped attributes. For example, if we want to build a model that predict the `median house value`, the Machine Learning algorithms may learn that prices never go beyond that limit. If your system needs precise predictions even beyond 500,000, then you have mainly two options:  
  1. Collect proper values for the districts whose labels were capped.
  2. Remove those districts from the training set (and also from the test set, since your system should not be evaluated poorly if it predicts values beyond $500,000).

### Boxplot

### Scatter plot

In the housing dataset, there is geographical information. Hence, we can create a scatterplot of all districts to visualize the data.

```python
housing.plot(kind="scatter", x="longitude", y="latitude")
plt.show()
```

The above plot is similar to the map of california which means that the collected districts are distributed in all the state. We can focus more on the places where there is a high density of data points using the parameter `alpha` which control the transparency of the points based on the spatial density of nearby points i.e. the areas with more crowded points will have darker color.
Other methods to show the density:  
https://stackoverflow.com/questions/49662964/density-scatter-plot-for-huge-dataset-in-matplotlib 
https://www.semicolonworld.com/question/53744/how-can-i-make-a-scatter-plot-colored-by-density-in-matplotlib 

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.1)
plt.savefig('scatter1.png')
plt.show()
```

Now let’s look at the housing prices. The radius of each circle represents the district’s population (option s), and the color represents the price (option c). We will use a predefined color map (option cmap) called jet, which ranges from blue (low values) to red (high prices)

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.1,
s=housing["population"]/100, label="population", figsize=(10,7),
c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True)
plt.show()
```

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
s=housing["population"]/100,label="population" , figsize=(15,10),
c="median_house_value", colormap=plt.get_cmap("jet"), colorbar=True)
plt.show()
```

This image tells you that the housing prices are very much related to the location (e.g., close to the ocean) and to the population density.

# 4 -  Data correlations

We can compute the Pearson's correlation coefficient between every pair of attributes using the `corr()` method

```python
corr_matrix=housing.corr()
```

```python
corr_matrix
```

Sometimes its good to plot the correlation matrix to visualize the relation between the attributes. We can use one of the following three methods. 

```python
plt.matshow(corr_matrix)
plt.colorbar()
plt.show()
```

```python
corr_matrix.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
import seaborn as sns
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix,annot=True)
```

Now let’s look at how much each attribute correlates with the `median house value`:

```python
corr_matrix["median_house_value"].sort_values(ascending=False)
```

Looking at the above numbers, we can see that the `median_house_value` has positive relation with the `median_income` and it has very weak relation to the other attributes.  
We can see a small negative correlation between the latitude and the median house value (i.e., prices have a slight tendency to go down when you go north.

Another way to check for correlation between attributes is to use Pandas’ `scatter_matrix` function, which plots every numerical attribute against every other numerical attribute. If there are 9 numerical attributes, you would get 9^2 = 81 plots, which would not fit on a page, so let’s just focus on a few promising attributes that seem most correlated with the median housing value.

```python

from pandas.plotting import scatter_matrix
attributes = ["median_house_value", "median_income", "total_rooms", "housing_median_age"]
scatter_matrix(housing[attributes], figsize=(12, 8))
plt.show()
```

The main diagonal (top left to bottom right) would be full of straight lines if Pandas plotted each variable against itself, which would not be very useful. So instead Pandas displays a histogram of each attribute (other options are available; see Pandas’ documentation for more details).  
The most promising attribute to predict the median house value is the median income, so let’s zoom in on their correlation scatterplot.

```python
housing.plot(kind="scatter", x="median_income", y="median_house_value", alpha=0.1)
```

This plot reveals a few things. First, the correlation is indeed very strong; you can clearly see the upward trend and the points are not too dispersed. Second, the price cap that we noticed earlier is clearly visible as a horizontal line at 500,000.
