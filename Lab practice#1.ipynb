Solve the example of Lecture 2-2, slide 9 to plot the quantile plot.

```python
x =[45, 50, 75,61, 90, 95, 99]
x.sort()  #only if not sorted

f=[]
for i in range(1,8):
    f.append((i -(0.5))/7)

#complete the missinng part (Find F)

%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(f,x)
plt.title('Quantile plot')
plt.xlabel('f value')
plt.ylabel('Scores')
plt.show()
print(f)
```
